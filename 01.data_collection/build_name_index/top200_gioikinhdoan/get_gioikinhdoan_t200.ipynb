```python
import json
import requests
import pandas as pd
from matplotlib import pyplot as plt
```

```python
query="""
    query
    {
        richestCeo(year: 2023, top: 200, gender: "A") {
            ceoCode
            currentOrder
            ceoImage
            ceoName
            ceoAge
            ceoCountry
        }
    }
"""
```

```python
json_data = {
    'query': query,
}
```

```python
response = requests.post('https://msh-data.cafef.vn/graphql/', json=json_data)
```

```python
response.json()
```

```python
df = pd.DataFrame(response.json()['data']['richestCeo'])
df.head()
```

```python
# drop ceoCode column
df.drop(columns=['ceoCode'], inplace=True)
# deduplicate
df.drop_duplicates(inplace=True)
```

```python
# remove rows with ceoImage is null or empty
df = df[df['ceoImage'].notnull()]
df = df[df['ceoImage'] != '']
# remove ceoImage column and currentOrder column
df.drop(columns=['ceoImage', 'currentOrder'], inplace=True)
df.shape
```

```python
# rename columns to ['name', 'province', 'age']
df.rename(columns={'ceoName': 'name', 'ceoCountry': 'province', 'ceoAge': 'age'}, inplace=True)
df.head()
```

```python
df.to_csv('kinh_doanh.csv', index=False)
```
